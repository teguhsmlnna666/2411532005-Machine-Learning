```python
def select_second(L):
    """Return the second element of the given list. If the list has no second
    element, return None.
    """
    if len(L) < 2:
        return None
    return L[1]

print(select_second([1,2,3]))
```

```python
def losing_team_captain(teams):
    """Given a list of teams, where each team is a list of names, return the 2nd player (captain)
    from the last listed team
    """
    return teams[-1][1]

teams = [
    ["Captain A","Player A"],["Captain B","Player B"],["Captain C","Player C"]
]

print(losing_team_captain(teams))
```

```python
def purple_shell(racers):
    """Given a list of racers, set the first place racer (at the front of the list) to last
    place and vice versa.

    >>> r = ["Mario", "Bowser", "Luigi"]
    >>> purple_shell(r)
    >>> r
    ["Luigi", "Bowser", "Mario"]
    """
    racers[0], racers[-1] = racers[-1], racers[0]

r = ["Mario", "Bowser", "Luigi"]
purple_shell(r)
print(r)
```

```python
def fashionably_late(arrivals, name):
    """Given an ordered list of arrivals to the party and a name, return whether the guest with that
    name was fashionably late.
    """
    index = arrivals.index(name)
    half = len(arrivals) / 2
    return index > half and index != len(arrivals) - 1

party_attendees = ['Adela', 'Fleda', 'Owen', 'May', 'Mona', 'Gilbert', 'Ford']

print(fashionably_late(party_attendees, 'Mona'))
print(fashionably_late(party_attendees, 'Gilbert'))
print(fashionably_late(party_attendees, 'Ford'))
```
