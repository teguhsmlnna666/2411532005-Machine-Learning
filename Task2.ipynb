```python
def round_to_two_places(num):
    """Return the given number rounded to two decimal places.

    >>> round_to_two_places(3.14159)
    3.14
    """
    # Replace this body with your own code.
    return round(num, 2)
    # ("pass" is a keyword that does literally nothing. We used it as a placeholder
    # because after we begin a code block, Python requires at least one line of code)

result = round_to_two_places(3.14159)

print ("hasil pembulatan =", result)
```

```python
x = 9876.543

# Pembulatan ke puluhan
hasil1 = round(x, -1)
# Pembulatan ke ratusan
hasil2 = round(x, -2)
# Pembulatan ke ribuan
hasil3 = round(x, -3)

print("Pembulatan ke puluhan:", hasil1)
print("Pembulatan ke ratusan:", hasil2)
print("Pembulatan ke ribuan:", hasil3)
```

```python
def to_smash(total_candies):
    """Return the number of leftover candies that must be smashed after distributing
    the given number of candies evenly between 3 friends.

    >>> to_smash(91)
    1
    """
    return total_candies % 3

print(to_smash(91))
```
